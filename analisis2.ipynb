```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar la base de datos Titanic desde Seaborn
df = sns.load_dataset("titanic")
```

```python
plt.figure(figsize=(8, 5))
sns.violinplot(x="pclass", y="fare", data=df, inner="quartile")
plt.title("Distribuci√≥n de Fare por Clase")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(data=df, x="age", hue="pclass", fill=True, common_norm=False, alpha=0.4)
plt.title("Densidad de Edad por Clase")
plt.show()
```

```python
cross_tab = pd.crosstab(df["sex"], df["class"])
cross_tab.plot(kind="bar", stacked=True, colormap="Pastel1", figsize=(8, 5))
plt.title("Distribuci√≥n de Pasajeros por Sexo y Clase")
plt.ylabel("N√∫mero de Pasajeros")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.regplot(x="age", y="survived", data=df, logistic=True, ci=95)
plt.title("Probabilidad de Supervivencia en funci√≥n de la Edad")
plt.show()
```
