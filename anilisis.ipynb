```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar la base de datos Titanic desde Seaborn
df = sns.load_dataset("titanic")
```

```python
# Punto 1: Resumen Estadístico
# 1. Relación entre pclass y fare
plt.figure(figsize=(8, 5))
sns.boxplot(x="pclass", y="fare", data=df)
plt.title("Distribución de Fare por Clase")
plt.show()
```

```python
# 2. Histograma de edades agrupadas por clase
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x="age", hue="pclass", kde=True, bins=20)
plt.title("Distribución de Edad por Clase")
plt.show()
```

```python
# 3. Gráfica de una variable categórica (Sexo)
plt.figure(figsize=(6, 4))
sns.countplot(x="sex", data=df, palette="pastel")
plt.title("Distribución de Pasajeros por Sexo")
plt.show()
```

```python
# Punto 2: Probabilidad de Supervivencia por Edad
menores_18 = df[df["age"] < 18]
probabilidad_supervivencia = menores_18["survived"].mean()
```

```python
print(f"La probabilidad de que un pasajero menor de 18 años haya sobrevivido es: {probabilidad_supervivencia:.2%}")
```
