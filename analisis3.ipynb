```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Carga db

```python
df = sns.load_dataset('titanic')
```

### Analizar relacion pclass y fare

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='pclass', y='fare', data=df)
plt.title('Distribución de tarifas por clase')
plt.xlabel('Clase del pasajero')
plt.ylabel('Tarifa')
plt.show()
```

### Histograma de edades agrupadas por clase

```python
plt.figure(figsize=(10, 6))
sns.histplot(df, x='age', hue='pclass', multiple='stack', bins=30, kde=True, palette='viridis')
plt.title('Distribución de edades por clase')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.legend(title='Clase')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

### Grafica variable categorica "sex" y supervivencia

```python
sns.countplot(x='sex', data=df, hue='survived')
plt.title('Distribución de pasajeros por género y supervivencia')
plt.xlabel('Género')
plt.ylabel('Cantidad')
plt.legend(title='Supervivencia', labels=['No', 'Sí'])
plt.show()
```

### 2. Probabilidad de supervivencia menores de 18

```python
menores_18 = df[df['age'] < 18]
```

```python
probabilidad_supervivencia = menores_18['survived'].mean()
print(f'La probabilidad de que un pasajero menor de 18 años haya sobrevivido es: {probabilidad_supervivencia:.2%}')
```
